```python
import pandas as pd
data=pd.read_csv("/content/wine-clustering.csv",index_col='Alcohol')
```

```python
x=data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit_transform(x)
```

```python
import pandas as pd
Df = pd.DataFrame(pca , columns = ['PC1', 'PC2'])
```

```python
import matplotlib.pyplot as plt
x=Df['PC1']
y=Df['PC2']
plt.scatter(x,y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA-Dimensionality Reduction')
plt.show() 
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,max_iter=150, random_state=20).fit(Df)
pl=kmeans.cluster_centers_
pl
```

```python
labels=kmeans.labels_
labels
```

```python
plt.scatter(x,y,c=labels,cmap='rainbow')
plt.scatter(pl[0][0],pl[0][1],color='orange')
plt.scatter(pl[1][0],pl[1][1],color='black')
plt.scatter(pl[2][0],pl[2][1],color='green')
plt.scatter(pl[3][0],pl[3][1],color='blue')
plt.show() 
```
